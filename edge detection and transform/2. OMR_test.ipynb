```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import scipy
from imutils import contours
from imutils.perspective import four_point_transform
```

```python
from cProfile import label


image = cv2.imread("./omr_test_01.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
more_blurred = cv2.GaussianBlur(gray, (9,9), 10,10)
less_blurred = cv2.GaussianBlur(gray, (9,9), 0.5,0.5)


## show how the function "gaussianBlur" works as sigma changes.

#more_blurred
mu_x, sigma_x = 0, 10
mx = np.linspace(mu_x-3*sigma_x, mu_x+3*sigma_x, 50)
my = scipy.stats.norm.pdf(mx, mu_x, sigma_x)
#less_blurred
mu_x, sigma_x = 0, 0.5
lx = np.linspace(mu_x-3*sigma_x, mu_x+3*sigma_x, 50)
ly = scipy.stats.norm.pdf(lx, mu_x, sigma_x)

plt.figure(figsize=(15,6))
plt.subplot(131)
plt.plot(mx, my, label="More Blurred (sigma = 10)")
plt.plot(lx,ly, label="Less Blurred (sigma=0.5)")
plt.title("Gaussian kernel shapes")
plt.legend(loc="upper right", prop={'size': 6})
plt.subplot(132)
plt.title("More Blurred Image")
plt.imshow(more_blurred, cmap="gray")
plt.subplot(133)
plt.title("Less Blurred Image")
plt.imshow(less_blurred, cmap="gray")
```
